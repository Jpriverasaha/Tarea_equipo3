```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st     
import pandas as pd
```

```python
## a)
fx = lambda x:(3/2)*(x ** 2) if -1 <= x <= 1 else 0
Fx = lambda x: x**(1/3) if x >= 0 else -(-x)**(1/3)
f_inv = lambda u : Fx((2*u)-1)
N = 1000
u1 = np.random.uniform(0,1,N)
muestras = list(map(lambda x : f_inv(x),u1))

y,x,p = plt.hist(muestras, bins=20, density=True)
y2 = list(map(lambda x: fx(x),x))
plt.plot(x,y2,'r')
plt.show()
```

```python
m1 = list(map(lambda x: f_inv(x), u1))
np.mean(m1)
```

```python
## b)

r1 = np.random.uniform(0,0.6,3)
r2 = np.random.uniform(0.6,0.9,3)
r3 = np.random.uniform(0.9,1,4)
y = [1/2,1,4]
r = [r1,r2,r3]

muestras = [list(map(lambda x: f_inv(x), i)) for i in r]
estrato = [list(map(lambda i, y: i/y, i,y))for i in muestras]
np.concatenate(estrato).mean()
```

```python
## c)

def estratos(x):
    U2 =np.random.rand(x)
    i =np.arange(0,x)
    estratos = (U2+i)/x
    return estratos
random = estratos(10)
muestras2 = list(map(lambda x: f_inv(x), random))
np.mean(muestras2)
```
